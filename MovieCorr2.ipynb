```python
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as npy

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\keegan.white\Downloads\movies1.csv')
```

```python
df.head()
```

```python
#looking for missing data
df = df.dropna()
# dropna removes cells that are null
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# datatypes in our columns
df.dtypes
```

```python
# removing the .0 from budgets, votes, and gross by changing datatypes
df['budget'] =df['budget'].astype('int64')
df['votes'] =df['votes'].astype('int64')
df['gross'] =df['gross'].astype('int64')
```

```python
df
```

```python
# the year and release date years are different so I'm changing the year to the release date year
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# Sorts movies by the gross sales
df.sort_values(by=['gross'], inplace = False, ascending = False) 
```

```python
#scatterplot showing the correlation between the budget and gross sales
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget Vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
df.head()
```

```python
# Regression plot using seaborn to see the correlation between budget and gross sales
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={"color": "blue"})
```

```python
# looking at the actual correlation in numbers
# high correlation between budget and gross
df.corr()
```

```python
# Creating a heatmap to show how correlated certain movie features are
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#assigning numbers to strings so we can see the performance of genres, directors, etc.
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
```

```python
# Creating a heatmap with the new numerized data
correlation_matrix=df_numerized.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# More correlations in number
df_numerized.corr()
```

```python
# Correlations stacked ontop of eachother
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```
